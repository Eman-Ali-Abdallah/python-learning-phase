```python
#1

print("write your list \nWhen you finish enter (done)")
list_=[]
while True :
    num=input(">> ")
    list_.append(num)
    if num=="done" or num=="Done":
       list_.pop()
       break
print(list_)
list_.sort()
print("Ascending order" , list_)
list_.sort(reverse=True)
print("Descending order", list_)
```

```python
#2

def modify_list(lst):
    while True:
        num = input(">> ")

        if num.lower() == "done":
            break

        lst.append(int(num))

    lst.append(100)
    lst.insert(2, 999)

    if 50 in lst:
        lst.remove(50)  

    print(lst)

your_list=[]
modify_list(your_list)
```

```python
def remove_all_occurance(lst,value):
    i=0
    while i<len(lst):
        if lst[i]==value:
            del lst[i]
        else :
            i+=1
x=90
list_=[90,80,9,70,90,67,90,56,9]
remove_all_occurance(list_,x)
print(list_)
```
